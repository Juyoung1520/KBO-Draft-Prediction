```python
#목록에 해당하는 행으로 변수 만들기
mpg['size'] = np.where((mpg['category'] == 'compact') |
                      (mpg['category'] == 'subcompact')|
                      (mpg['category'] == '2seater'),
                      'small', 'large')
mpg['size'].value_counts()
```

```python
#위와 같이 |이 반복되는 조건문이라면, df.isin()을 사용하면 간편하게 작성할 수 있다. isin([]) 변수의 값이 입력한 목록에 해당하는지 확인하는 기능을 한다
mpg['size'] = np.where(mpg['category'].isin(['compact','subcompact','2seater']),'small','large')
mpg['size'].value_counts()
```

```python
#분석 도전 문제
midwest = pd.read_csv('midwest.csv')
midwest
```

```python
midwest = midwest.rename(columns = {'poptotal':'total','popasian':'asian'}) #변수 수정하기
midwest.head()
```

```python
midwest['percentage'] =( midwest['asian']/midwest['total'] ) * 100 # 전체 인구 대비 아시아 인구 백분율 구하기
midwest.head()
```

```python
midwest['percentage'].plot.hist() #히스토그램을 만들어 분포 살피기 -> 거의 대부분 도시의 아시아 인구 백분율이 1%도 되지 않는다. 
```

```python
#아시아 인구 백분율 전체 평균
mean = sum(midwest['percentage'])/len(midwest)
mean
```

```python
midwest['var1'] = np.where(midwest['percentage']>=mean, 'large', 'small') #아시아 인구 백분율 전체 평균을 초과하면 large, 그 외에는 small 부여
midwest.head()
```

```python
count_var1 = midwest['var1'].value_counts()
print(count_var1)
count_var1.plot.bar(rot = 0)
#'large'와 'small'에 해당하는 지역이 얼마나 많은지 빈도표와 빈도 막대 그래프를 만들어 확인해보기
```
