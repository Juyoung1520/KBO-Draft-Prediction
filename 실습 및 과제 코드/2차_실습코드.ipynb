```python
# 혼자서 해보기
# 01. 자동차 배기량에 따라 고속도로 연비가 다른지 알아보려고 한다. displ(배기량)이 4 이하인 자동차와 5 이상인 자동차 중에서 어떤 자동차의 hwy(고속도로 연비)평균이 더 높은지 알아보시오.
import pydataset as pds
df = pds.data('mpg')
df
displ_4 = df.query('displ <= 4')
displ_5 = df.query('displ >= 5')

avg_4 = displ_4['hwy'].mean()
avg_5 = displ_5['hwy'].mean()

print(avg_4, avg_5)

#배기량이 4 이하인 자동차가 평균 고속도로 연비가 더 높다
```
