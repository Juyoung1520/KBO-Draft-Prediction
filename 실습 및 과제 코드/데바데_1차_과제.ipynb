# 2. 사칙연산 함수

### 함수부분

```python
def add(a, c):
    return a + c

def subtract(a, c):
    return a - c

def multiply(a, c):
    return a * c

def divide(a, c):
    if c == 0:
        return "False"
    return a / c
```

### 입력받기

```python
a, b, c = input().split()

a = int(a)
c = int(c)

if b not in ['+', '-', '*', '/']:
    result = "False"
else:
    if b == '+':
        result = add(a, c)
    elif b == '-':
        result = subtract(a, c)
    elif b == '*':
        result = multiply(a, c)
    elif b == '/':
        result = divide(a, c)


if isinstance(result, float):
    print(int(result))  # 정수 부분만 출력
else:
    print(result)
```
