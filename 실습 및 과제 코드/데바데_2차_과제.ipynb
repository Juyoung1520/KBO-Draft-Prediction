```python
#전처리 사용자 함수 정의 
def text_preprocessor(s):
    import re
    
    #괄호, 괄호 안 문자 제거햐기
    pattern = r'\([^)]*\)'
    s = re.sub(pattern = pattern, repl='', string=s)
    
    pattern = r'\[[^)]*\]'
    s = re.sub(pattern = pattern, repl='', string = s)
    
    pattern = r'\<[^)]*\>'
    s = re.sub(pattern = pattern, repl='', string = s)
    
    pattern = r'\{[^)]*\}'
    s = re.sub(pattern = pattern, repl='', string = s)
    
    
    #특수문자 제거하기
    pattern = r'[^a-zA-Z가-힣]'
    s = re.sub(pattern = pattern, repl=' ', string = s)
    
    #단위 제거하기
    units = ['mm', 'cm', 'm', 'km', 'ml', 'kg', 'g', 'mg', '㎡']
    for unit in units:
        s = s.replace(unit,' ')
   
        
    #공백 기준으로 분할
    s_split = s.split()
 
    
    #한 글자만 있는 단어 제외
    s_list = []
    for word in s_split:
        if len(word) != 1:
            s_list.append(word)
    
    return s_list
article_pre = text_preprocessor(article.text)
print(article_pre )
```

```python
#문장의 평균 길이를 계산하시오.
#문장은 온점 단위로 나눌 수 있다. 
strings = article.text.split('.')

A = []
for s in strings:
    pattern = r'[^a-zA-z가-힣]'
    s2 = re.sub(pattern = pattern, repl = ' ', string = s)
    A.append(s2)
print(A) #특수문자 삭제한 문장들로 이루어진 리스트
    
total = 0
n = len(A)
for i in A:
    length = len(i.split(' '))
    total += length
total/n #문장의 평균 길이 (단어 개수)
```

```python
#감정 분석
#한글 감성어 사전 업로드
import json
import pandas as pd

with open('SentiWord_info.json', encoding='utf-8-sig', mode= 'r') as f :
    SentiWord_info = json.load(f)
sentiword_dic = pd.DataFrame(SentiWord_info)

print(sentiword_dic)


#긍정적인 목록, 부정적인 목록 정의하기
pos = []
neg = []

for i in range(len(sentiword_dic)):
    if int(sentiword_dic['polarity'][i]) >= 1: #극성이 1 이상은 긍정적인 단어로 판단
        pos.append(sentiword_dic['word_root'][i])
    elif int(sentiword_dic['polarity'][i]) <= -1: #극성이 -1 이하는 부정적인 단어로 판단
        neg.append(sentiword_dic['word_root'][i])

# 긍정적인 단어와 부정적인 단어의 빈도 계산
pos_score = 0
neg_score = 0
for token in article_pre : #전처리 완료한 자료 활용
    if token in pos: # 토큰이 긍정적인 목록에 포함된 단어라면 긍정적인 단어의 빈도 점수 +1
        pos_score += 1
        print(token) # 무슨 단어인지 확인
    elif token in neg: # 토큰이 부정적인 목록에 포함된 단어라면 부정적인 단어의 빈도 점수 +1
        neg_score += 1
        print(token)
    
print('긍정적인 단어의 빈도',pos_score)
print('부정적인 단어의 빈도',neg_score)

#긍적적인 단어는 3번, 부정적인 단어는 4번 나옴을 알 수 있다.
```
