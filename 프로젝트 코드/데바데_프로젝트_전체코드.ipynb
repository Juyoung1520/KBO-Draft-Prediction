```python
#type 변경
df[['RB','RR','RL','LL','LR']] = df[['RB', 'RR', 'RL', 'LL', 'LR']].astype(int) #boolean type -> int type

#설명변수 설정
x_data_backward = df[['height',
       'weight', 'ERA', 'G', 'W', 'L', 'WPCT', 'hitter', 'AB', 'pitches', 'IP',
       'Hits', 'homerun', 'SH', 'SF', 'FB', 'IBB', 'BB', 'SO', 'wild_pitch',
       'balk', 'point', 'earned_run', 'WHIP', 'AVG_h', 'AVG_s', 'RB', 'RR','RL','LL', 'LR']]

#상수항 추가 및 종속변수 설정
x_data_backward = sm.add_constant(x_data_backward)
X_backward = x_data_backward
y_backward = df['nomination']

#데이터셋 분리
X_train_, X_test_, y_train_, y_test_ = train_test_split(X_backward,y_backward,stratify = y_backward,train_size = 0.7, test_size = 0.3
                                                    , random_state = 1 )
print(X_train_.shape, X_test_.shape, y_train_.shape, y_test_.shape)

X_test_
```

### 2.1.2. Backward Selection을 이용한 변수 선택

```python
# 변수 선택법 함수

def processSubset(X,y,feature_set):
    model = sm.OLS(y,X[list(feature_set)]) #modeling
    regr = model.fit()
    AIC = regr.aic #모델의 AIC
    return {'model' : regr, 'AIC': AIC}



print(processSubset(X = X_train_, y = y_train_, feature_set = X_backward.columns))
```

```python
#후진 선택법

import itertools
import time

def backward(X,y,predictors) : 
    tic = time.time() 
    results = []
    # 데이터 변수들이 미리 정의된 predictors 조합 확인
    for combo in itertools.combinations(predictors, len(predictors)-1):
        results.append(processSubset(X=X, y=y, feature_set = list(combo)+['const']))
    models = pd.DataFrame(results)
    
    #가장 낮은 AIC를 가진 모델 선택
    
    best_model = models.loc[models['AIC'].argmin()]
    toc = time.time()
    print("Processed ", models.shape[0], "models on", len(predictors) - 1,"predictors in",(toc-tic))
    print('Selected predictors: ',best_model['model'].model.exog_names,' AIC:', best_model[0])
    return best_model
```

```python
#후진 소거법 모델

def backward_model(X,y):
    Bmodels = pd.DataFrame(columns = ["AIC", "model"], index = range(1,len(X_backward.columns)))
    tic = time.time()
    
    predictors = X_backward.columns.difference(['const'])
    
    Bmodel_before = processSubset(X,y,predictors)['AIC']
    
    while (len(predictors)>1):
        Backward_result = backward(X = X_train_, y = y_train_, predictors = predictors)
        if Backward_result['AIC'] > Bmodel_before :
            break
        Bmodels.loc[len(predictors)-1] = Backward_result
        predictors = Bmodels.loc[len(predictors)-1]["model"].model.exog_names
        Bmodel_before = Backward_result['AIC']
        predictors = [k for k in predictors if k!= 'const']
        
    toc = time.time()
    print('total elapsed time:', (toc-tic), 'seconds.')
    return (Bmodels['model'].dropna().iloc[0])
```

AIC 가 작아질 때까지 좋은 조합의 features를 출력하다가 AIC가 다시 커질 때 while문이 작동하지 않으므로 결과적으로 가장 마지막에 출력된 features가 최적의 설명변수다.

```python
Backward_best_model = backward_model(X=X_train_, y = y_train_)
```
